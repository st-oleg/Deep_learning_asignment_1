# 3 Simple Classification

A netowrk with 1 softmax layer is used. The loss funciton is categorical cross-entropy

With SGD optimizer the model does not converge in 200 epochs, the accuracy and loos funciton seem to stay the same as at step1, after random weight initialization
With adam optimizer the model converges much faster, reaching 100% accuracy after 20 epochs

```python
# Generating trainig data
# 0 - rectangle, 1- disc, 2 - triangle

[X_train, Y_train] = generate_dataset_classification(300, 20)
```

```python
# Linear classifier

n_classes = 3
model =Sequential()
model.add(Dense(n_classes,input_dim=np.shape(X_train)[1],activation='softmax'))
model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
model.fit(X_train,to_categorical(Y_train),epochs=100,batch_size=32)
```

```python
# testing on triabgles

X_test = generate_a_triangle()
X_test = X_test[0].reshape(1, X_test[0].shape[0])
print(model.predict(X_test))

plt.imshow(X_test.reshape(IMAGE_SIZE,IMAGE_SIZE), cmap='gray')
```

```python
# testing on disc

X_test = generate_a_disk()
X_test = X_test.reshape(1, X_test.shape[0])
print(model.predict(X_test))

plt.imshow(X_test.reshape(IMAGE_SIZE,IMAGE_SIZE), cmap='gray')
```

```python
# testing on rectangle

X_test = generate_a_rectangle()
X_test = X_test.reshape(1, X_test.shape[0])
print(model.predict(X_test))

plt.imshow(X_test.reshape(IMAGE_SIZE,IMAGE_SIZE), cmap='gray')
```

# 5 A More Difficult Classification Problem

On this problem the linear classifier seems to converge much slower. 
However after ~500 epochs the it reaches the 100% train accuracy
The test accuracy is much lower than train - 60%, 

```python
# Generating new data
[X_train, Y_train] = generate_dataset_classification(300, 20, True)
[X_test, Y_test] = generate_test_set_classification()
```

```python
# Retrainig the linear classifier

n_classes = 3
model =Sequential()
model.add(Dense(n_classes,input_dim=np.shape(X_train)[1],activation='softmax'))
model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
model.fit(X_train,to_categorical(Y_train),epochs=1000,batch_size=32)
```

# Convolutional network

The convolutional network with 1 convolutional layer with 16 5x5 filters shows fast convergence and reaches 100% train accuracy in 34 epochs. The test accuracy is also improved - 72%

```python
# Transformig data for convoluitonal layer
# The pictures should be represented by matrixes not vectors

X_train_conv = np.reshape(X_train,(300,72,72,1))
X_test_conv = np.reshape(X_test,(300,72,72,1))
```

```python
# Trainng a convolutional neural network

n_classes = 3
model_conv =Sequential()
# convolutinal layer
model_conv.add(Conv2D(16,(5,5),input_shape=X_train_conv.shape[1:],activation='relu'))
# pooling layer
model_conv.add(MaxPooling2D())
# output layer
model_conv.add(Flatten())
model_conv.add(Dense(n_classes,activation='softmax'))

# compiling and fitting
model_conv.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
model_conv.fit(X_train_conv,to_categorical(Y_train),epochs=50,batch_size=32)
```

```python
# Testing 
model_conv.evaluate(X_test_conv, Y_test)
```

# 6 A Regression Problem

The simple linear model adapted for the regerssion task does not perform well, the visualized predicitond do not make sense.
Adding an extra dense layer did not improve the result, but the traing time increased

The convoluitonal model with 1 conv layer and output with relu activation for last layer outputs zeros, the performance is better with the last layer with no activation funciton 

The model overfits after ~10 epochs, the train error drops but the test error starts to grow, predicitons on train look accurate when visualized but test ones dont.

Also an extra convolutional layer seems to be less computationly expensive, which is theoreticly correct - there are less parameters to learn


By adding more convolutional layers and a dropout layer the overfitting problem disappered, the convergence seems more steady and train accuracy is close to test, the best score obtained is ~0.7 mean square error. 

Finally, a reduced U-net network was tested. For these task a reduction of dimensionality is neede, so the second half of the network was removed. It performed similar to the previous conv-net





```python
# Generating data
[X_train, Y_train] = generate_dataset_regression(300, 20)
[X_test, Y_test] = generate_test_set_regression()
```

```python
# Feed forward network
regressor = Sequential()
regressor.add(Dense(np.shape(X_train)[1],input_dim=np.shape(X_train)[1],activation='linear'))
regressor.add(Dense(np.shape(Y_train)[1],activation='linear'))
regressor.compile(optimizer='adam',
                  loss='mean_squared_error')
regressor.fit(X_train,Y_train,epochs=20,batch_size=32,validation_data=(X_test,Y_test))
```

```python
# convoluitonal network

X_train_conv = np.reshape(X_train,(300,72,72,1))
X_test_conv = np.reshape(X_test,(300,72,72,1))


model_conv =Sequential()
# convolutinal layer
model_conv.add(Conv2D(16,(5,5),input_shape=X_train_conv.shape[1:],activation='relu'))
# pooling layer
model_conv.add(MaxPooling2D())
# output layer
model_conv.add(Flatten())
model_conv.add(Dense(6))

# compiling and fitting
model_conv.compile(optimizer='adam',
                  loss='mean_squared_error')
model_conv.fit(X_train_conv,Y_train,epochs=50,batch_size=32,validation_data=(X_test_conv,Y_test))
```

```python
# test
model_conv.evaluate(X_test_conv, Y_test)
```

```python
# Visualize predictions

x_var = X_test_conv
y_var = Y_test
n_example = 25

Y_pred = model_conv.predict(x_var)
visualize_prediction(x_var[n_example], Y_pred[n_example])
```

```python
# convoluitonal network with more layers

X_train_conv = np.reshape(X_train,(300,72,72,1))
X_test_conv = np.reshape(X_test,(300,72,72,1))


model_conv =Sequential()
# convolutinal layer
model_conv.add(Conv2D(16,(5,5),input_shape=X_train_conv.shape[1:],activation='relu'))
model_conv.add(Conv2D(3,(5,5),activation='relu'))
model_conv.add(MaxPooling2D())

model_conv.add(Conv2D(5,(3,3),activation='relu'))
model_conv.add(Conv2D(3,(3,3),activation='relu'))
model_conv.add(MaxPooling2D())

model_conv.add(Dropout(0.5))

# output layer
model_conv.add(Flatten())
model_conv.add(Dense(6))

# compiling and fitting
model_conv.compile(optimizer='adam',
                  loss='mean_squared_error')
model_conv.fit(X_train_conv,Y_train,epochs=50,batch_size=32,validation_data=(X_test_conv,Y_test))
```

```python
# Visualize predictions

x_var = X_train_conv
y_var = Y_train
n_example = 4

Y_pred = model_conv.predict(x_var)
visualize_prediction(x_var[n_example], Y_pred[n_example])
```

```python
# 1/2 of U-net 

X_train_conv = np.reshape(X_train,(300,72,72,1))
X_test_conv = np.reshape(X_test,(300,72,72,1))


model_conv =Sequential()
# convolutinal layer
model_conv.add(Conv2D(64,(3,3),input_shape=X_train_conv.shape[1:],activation='relu'))
model_conv.add(Conv2D(64,(3,3),activation='relu'))
model_conv.add(MaxPooling2D(pool_size=(2,2)))

model_conv.add(Conv2D(128,(3,3),activation='relu'))
model_conv.add(Conv2D(128,(3,3),activation='relu'))
model_conv.add(MaxPooling2D(pool_size=(2,2)))

model_conv.add(Conv2D(256,(3,3),activation='relu'))
model_conv.add(Conv2D(256,(3,3),activation='relu'))
model_conv.add(MaxPooling2D(pool_size=(2,2)))

model_conv.add(Dropout(0.5))

# output layer
model_conv.add(Flatten())
model_conv.add(Dense(5184,activation='relu'))
model_conv.add(Dense(6))

# compiling and fitting
model_conv.compile(optimizer='adam',
                  loss='mean_squared_error')
model_conv.fit(X_train_conv,Y_train,epochs=50,batch_size=32,validation_data=(X_test_conv,Y_test))
```

# 7 Image Denoising
For this taks, the generate_dataset_denoise funciton was created, similar to the generate_dataset_classification
Also functions generate_a_triangle,(disc,rectangle) were modified to give a dual output - image with and without noise. These funcitons are in the starter_code.py

A hour-glass network was built, as described in the lecture
The number of layers was chosen to keep the training time reasonable.



```python
# Generate data, split to trian and test
[Y,X] = generate_dataset_denoise(300,100)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
# visualise train examples
# with noise
plt.imshow(X[0].reshape(IMAGE_SIZE,IMAGE_SIZE), cmap='gray')
```

```python
# without noise
plt.imshow(Y[0].reshape(IMAGE_SIZE,IMAGE_SIZE), cmap='gray')
```

```python
# Hourglass Network

X_train_conv = np.reshape(X_train,(len(X_train),72,72,1))
X_test_conv = np.reshape(X_test,(len(X_test),72,72,1))


model_conv =Sequential()
# convolutinal layer
model_conv.add(Conv2D(64,(3,3),input_shape=X_train_conv.shape[1:],activation='relu'))
model_conv.add(BatchNormalization())
model_conv.add(Conv2D(64,(3,3),activation='relu'))
model_conv.add(BatchNormalization())
model_conv.add(MaxPooling2D(pool_size=(2,2)))

model_conv.add(Conv2D(128,(3,3),activation='relu'))
model_conv.add(BatchNormalization())
model_conv.add(Conv2D(128,(3,3),activation='relu'))
model_conv.add(BatchNormalization())
model_conv.add(MaxPooling2D(pool_size=(2,2)))

model_conv.add(Conv2D(256,(3,3),activation='relu'))
model_conv.add(BatchNormalization())
model_conv.add(Conv2D(256,(3,3),activation='relu'))
model_conv.add(BatchNormalization())
model_conv.add(Conv2D(256,(3,3),activation='relu'))
model_conv.add(BatchNormalization())
model_conv.add(MaxPooling2D(pool_size=(2,2)))

model_conv.add(UpSampling2D(size = (2,2)))
model_conv.add(Conv2D(256, 2, activation = 'relu'))
model_conv.add(BatchNormalization())
model_conv.add(Conv2D(256, 3, activation = 'relu'))
model_conv.add(BatchNormalization())
model_conv.add(Conv2D(256, 3, activation = 'relu'))
model_conv.add(BatchNormalization())

model_conv.add(UpSampling2D(size = (2,2)))
model_conv.add(Conv2D(128, 2, activation = 'relu'))
model_conv.add(BatchNormalization())
model_conv.add(Conv2D(128, 3, activation = 'relu'))
model_conv.add(BatchNormalization())
# model_conv.add(Conv2D(128, 3, activation = 'relu'))

model_conv.add(UpSampling2D(size = (2,2)))
model_conv.add(Conv2D(64, 2, activation = 'relu'))
model_conv.add(Conv2D(64, 3, activation = 'relu'))
# model_conv.add(Conv2D(64, 3, activation = 'relu'))



# model_conv.add(Dropout(0.5))

# output layer
model_conv.add(Flatten())
model_conv.add(Dense(5184,activation='softmax'))

# compiling and fitting
model_conv.compile(optimizer='adam',
                  loss='mean_squared_error')
model_conv.fit(X_train_conv,Y_train,epochs=20,batch_size=32,validation_data=(X_test_conv,Y_test))
```
